```python
import re


class Solution:
    def numberToWords(self, num: int) -> str:
        if num == 0:
            return "Zero"
        
        billions = num // 1_000_000_000
        millions = (num % 1_000_000_000) // 1_000_000
        thousands = (num % 1_000_000) // 1_000
        hundreds = num % 1_000

        result = ""
        if billions > 0:
            # print(f"billions: {billions}")
            result += self.get_digit_hundreds(billions) + " Billion"

        if millions > 0:
            # print(f"millions: {millions}")
            # print(f"result before millions: '{result}'")
            if result != "":
                result += " "
            result += self.get_digit_hundreds(millions) + " Million"

        if thousands > 0:
            # print(f"thousands: {thousands}")
            # print(f"result before thousands: '{result}'")
            if result != "":
                result += " "
            result += self.get_digit_hundreds(thousands) + " Thousand"

        if hundreds > 0:
            # print(f"hundreds: {hundreds}")
            # print(f"result before hundreds: '{result}'")
            if result != "":
                result += " "
            result += self.get_digit_hundreds(hundreds)
        
        return result
        
    def get_digit_simple(self, digit: int) -> str:
        if digit == 1:
            return "One"
        if digit == 2:
            return "Two"
        if digit == 3:
            return "Three"
        if digit == 4: 
            return "Four"
        if digit == 5:
            return "Five"
        if digit == 6:
            return "Six"
        if digit == 7:
            return "Seven"
        if digit == 8:  
            return "Eight"
        if digit == 9:  
            return "Nine"
        return ""

    def get_digit_tens(self, digits: int) -> str:
        if digits < 10:
            return self.get_digit_simple(digits)
        if digits == 10:
            return "Ten"
        if digits == 11:
            return "Eleven"
        if digits == 12:
            return "Twelve"
        if digits == 13:
            return "Thirteen"
        if digits == 14:
            return "Fourteen"
        if digits == 15:
            return "Fifteen"
        if digits == 16:
            return "Sixteen"
        if digits == 17:
            return "Seventeen"
        if digits == 18:
            return "Eighteen"
        if digits == 19:
            return "Nineteen"
        if digits // 10 == 2:
            return "Twenty" + (" " + self.get_digit_simple(digits % 10) if digits % 10 != 0 else "")
        if digits // 10 == 3:
            return "Thirty" + (" " + self.get_digit_simple(digits % 10) if digits % 10 != 0 else "")
        if digits // 10 == 4:
            return "Forty" + (" " + self.get_digit_simple(digits % 10) if digits % 10 != 0 else "")
        if digits // 10 == 5:
            return "Fifty" + (" " + self.get_digit_simple(digits % 10) if digits % 10 != 0 else "")
        if digits // 10 == 6:
            return "Sixty" + (" " + self.get_digit_simple(digits % 10) if digits % 10 != 0 else "")
        if digits // 10 == 7:
            return "Seventy" + (" " + self.get_digit_simple(digits % 10) if digits % 10 != 0 else "")
        if digits // 10 == 8:
            return "Eighty" + (" " + self.get_digit_simple(digits % 10) if digits % 10 != 0 else "")
        if digits // 10 == 9:
            return "Ninety" + (" " + self.get_digit_simple(digits % 10) if digits % 10 != 0 else "")
        return ""
    
    def get_digit_hundreds(self, digits: int) -> str:
        hundreds = digits // 100
        tens = digits % 100
        result = ""

        if hundreds > 0:
            result += self.get_digit_simple(hundreds) + " Hundred" + (" " + self.get_digit_tens(tens) if tens != 0 else "")
        else:
            result += self.get_digit_tens(tens)

        return result
    
        
```

```python
soln = Solution()

num = 123
expected = "One Hundred Twenty Three"
output = soln.numberToWords(num)
print(f"output: {output}, expected: {expected}, pass: {output == expected}")

num = 12345
expected = "Twelve Thousand Three Hundred Forty Five"
output = soln.numberToWords(num)
print(f"output: {output}, expected: {expected}, pass: {output == expected}")

num = 1234567
expected = "One Million Two Hundred Thirty Four Thousand Five Hundred Sixty Seven"
output = soln.numberToWords(num)
print(f"output: {output}, expected: {expected}, pass: {output == expected}")

num = 11234567
expected = "Eleven Million Two Hundred Thirty Four Thousand Five Hundred Sixty Seven"
output = soln.numberToWords(num)
print(f"output: {output}, expected: {expected}, pass: {output == expected}")
```

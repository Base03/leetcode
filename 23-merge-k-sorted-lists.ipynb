```python
from typing import List, Optional

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

    # print linked list for debugging
    def __str__(self):
        result = []
        current = self
        while current:
            result.append(str(current.val))
            current = current.next
        return ",".join(result)
        
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        if len(lists) == 0:
            return None
        if len(lists) == 1:
            return lists[0]
        if len(lists) == 2:
            return self.mergeLists(lists[0], lists[1])
        
        mid = len(lists) // 2
        left = self.mergeKLists(lists[:mid])
        right = self.mergeKLists(lists[mid:])
        return self.mergeLists(left, right)

    def mergeLists(self, a: Optional[ListNode], b: Optional[ListNode]) -> Optional[ListNode]:
        if not a:
            return b
        if not b:
            return a
        
        head = ListNode(0)
        current = head

        while a and b:
            if a.val < b.val:
                current.next = a
                a = a.next
            else:
                current.next = b
                b = b.next

            current = current.next
        
        if a:
            current.next = a
        if b:
            current.next = b

        return head.next
        
```

```python
soln = Solution()

list1 = ListNode(1, ListNode(4, ListNode(5)))
list2 = ListNode(1, ListNode(3, ListNode(4)))
list3 = ListNode(2, ListNode(6))
lists = [list1, list2, list3]
merged = soln.mergeKLists(lists)
expected = "1,1,2,3,4,4,5,6"
print(f"merged: {merged}, expected: {expected}, equal: {str(merged) == expected}")

lists = []
merged = soln.mergeKLists(lists)
expected = "None"
print(f"merged: {merged}, expected: {expected}, equal: {str(merged) == expected}")

    
```

```python

from typing import List, Tuple

class Solution:
    def calculateMinimumHP(self, dungeon: List[List[int]]) -> int:
        n = len(dungeon)
        m = len(dungeon[0])
        # negative of health required to reach the end from the current cell (inclusive)
        dp = [[float('-inf') for _ in range(m+1)] for _ in range(n+1)]

        dp[n][m-1] = -1
        dp[n-1][m] = -1
        # start at the princess and iterate backwards
        for i in range(n-1, -1, -1):
            for j in range(m-1, -1, -1):
                dp[i][j] = dungeon[i][j] + min(max(dp[i+1][j], dp[i][j+1]), -1)
        
        return -dp[0][0] if dp[0][0] < 0 else 1
    
    def calculateMinimumHP(self, dungeon: List[List[int]]) -> int:
        n: int = len(dungeon)
        m: int = len(dungeon[0])
        # health required to reach the end from the current cell (inclusive)
        dp: List[List[int]] = [[None for _ in range(m)] for _ in range(n)]
        ac: List[List[bool]] = [[False for _ in range(m)] for _ in range(n)]

        queue: List[Tuple[int,int]] = []

        def update_cells(i:int, j:int, health_needed:int) -> None:
            dp[i][j] = health_needed

            if i > 0 and not ac[i-1][j]:
                queue.append((i-1, j))
                ac[i-1][j] = True

            if j > 0 and not ac[i][j-1]:
                queue.append((i, j-1))
                ac[i][j-1] = True
        
        # require 1 health after entering the final cell
        i = n-1
        j = m-1
        health_needed = max(1, 1-dungeon[i][j])
        update_cells(i, j, health_needed)

        while queue:
            i, j = queue.pop(0)
            d_cost = float('inf') if i == n-1 else dp[i+1][j]
            r_cost = float('inf') if j == m-1 else dp[i][j+1]
            b_cost = min(d_cost, r_cost)
            
            if dungeon[i][j] >= 0:
                # heals, health required either 1 or b_cost + damage
                health_needed = max(1, b_cost - dungeon[i][j])
            else:
                # damages, health required b_cost - damage
                health_needed = b_cost - dungeon[i][j]
            
            update_cells(i, j, health_needed)
            
        return dp[0][0]
    
    def calculateMinimumHP(self, dungeon: List[List[int]]) -> int:
        n: int = len(dungeon)
        m: int = len(dungeon[0])
        # health required to reach the end from the current cell (inclusive)
        dp: List[List[int]] = [[float('inf') for _ in range(m+1)] for _ in range(n+1)]
        dp[n-1][m-1] = max(1, 1-dungeon[n-1][m-1])

        for i in range(n-1, -1, -1):
            for j in range(m-1, -1, -1):
                if i == n-1 and j == m-1:
                    continue

                d_cost = dp[i+1][j]
                r_cost = dp[i][j+1]
                b_cost = min(d_cost, r_cost)
                if dungeon[i][j] >= 0:
                    dp[i][j] = max(1, b_cost - dungeon[i][j])
                else:
                    dp[i][j] = b_cost - dungeon[i][j]
        
        return dp[0][0]
    
    def calculateMinimumHP(self, dungeon: List[List[int]]) -> int:
        n: int = len(dungeon)
        m: int = len(dungeon[0])
        # health required to reach the end from the current cell (inclusive)
        dp: List[List[int]] = [[float('inf') for _ in range(m+1)] for _ in range(n+1)]
        dp[n-1][m-1] = max(1, 1-dungeon[n-1][m-1])

        for i in range(n-1, -1, -1):
            for j in range(m-1, -1, -1):
                if i == n-1 and j == m-1:
                    continue

                val = min(dp[i+1][j], dp[i][j+1]) - dungeon[i][j]
                dp[i][j] = max(1, val) if dungeon[i][j] >= 0 else val
        
        return dp[0][0]
    
    def calculateMinimumHP(self, dungeon: List[List[int]]) -> int:
        n: int = len(dungeon)
        m: int = len(dungeon[0])
        # health required to reach the end from the current cell (inclusive)
        dp: List[List[int]] = [[float('inf') for _ in range(m+1)] for _ in range(n+1)]
        dp[n][m-1] = 1
        dp[n-1][m] = 1

        for i in range(n-1, -1, -1):
            for j in range(m-1, -1, -1):
                val = min(dp[i+1][j], dp[i][j+1]) - dungeon[i][j]
                dp[i][j] = max(1, val) if dungeon[i][j] >= 0 else val
        
        return dp[0][0]


        
```

```python
class Solution:
    def shortestPalindrome(self, s: str) -> str:
        n = len(s)
        # run over all possible odd-length palindromes
        min_delta = n
        for i in range(n):
            l, r = i,i
            while l >= 0 and r < n:
                if s[l] != s[r]:
                    break
                if l == 0:
                    min_delta = min(min_delta, n-r-1)
                l = l-1
                r = r+1

        for i in range(n-1):
            l, r = i, i+1
            while l >= 0 and r < n:
                if s[l] != s[r]:
                    break
                if l == 0:
                    min_delta = min(min_delta, n-r-1)
                l = l-1
                r = r+1

        print(f"min_delta: {min_delta}")
        print(f"s[n-min_delta:][::-1]: {s[n-min_delta:][::-1]}")
        return s[n-min_delta:][::-1] + s
        
```

```python
class Solution:
    def shortestPalindrome(self, s: str) -> str:
        n = len(s)
        if s == "":
            return ""

        for i in range(n,0,-1):
            prefix = s[i:][::-1]
            trial = prefix + s
            if trial == trial[::-1]:
                return trial
```

```python
class Solution:
    def shortestPalindrome(self, s: str) -> str:
        n: int = len(s)
        ss: str = s + "#" + s[::-1]
        nn = len(ss)
        table = [0]*nn

        for i in range(1,nn):
            j = table[i-1]

            while (j > 0 and ss[i] != ss[j]):
                j = table[j-1]
            
            if (ss[i] == ss[j]):
                j = j+1
            
            table[i] = j
        
        print(f"table: {table}")
```

```python
soln = Solution()

s = "abcd"
result = soln.shortestPalindrome(s)
expected = "abcddcba"
print(f"result: {result}, expected: {expected}, pass: {result == expected}")

s = "abcdefghijk"
result = soln.shortestPalindrome(s)
expected = "kjihgfedcbabcdefghijk"
print(f"result: {result}, expected: {expected}, pass: {result == expected}")

s = "aacecaaa"
result = soln.shortestPalindrome(s)
expected = "aaacecaaa"
print(f"result: {result}, expected: {expected}, pass: {result == expected}")

s = "aaacecaaa"
result = soln.shortestPalindrome(s)
expected = "aaacecaaa"
print(f"result: {result}, expected: {expected}, pass: {result == expected}")
```

```python
# don't give completions longer than a single line, I am doing this myself, 
# not relying on copilot

# recursive solution
class Solution(object):
    def climbStairs(self, n):
        """
        :type n: int
        :rtype: int
        """

        if n == 1:
            return 1
        if n == 2:
            return 2
        
        return self.climbStairs(n-1) + self.climbStairs(n-2)
```

```python
# don't give completions longer than a single line, I am doing this myself, 
# not relying on copilot
from typing import Dict

# recursive solution with memoization
class Solution(object):
    def climbStairsAux(self, n: int, memo: Dict[int, int]) -> int:
        if n == 0 or n == 1:
            return 1
        
        if n in memo:
            return memo[n]
        else:
            memo[n] = self.climbStairsAux(n-2, memo) + self.climbStairsAux(n-1, memo)

        return memo[n]

    def climbStairs(self, n: int) -> int:
        memo = {}
        return self.climbStairsAux(n, memo)
```

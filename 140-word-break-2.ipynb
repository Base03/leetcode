```python
from collections import defaultdict, deque
from typing import List, Dict

class Solution:
    def wordBreakRec(self, s: str, path: str, result: List[str], wordDict: List[str], start: int) -> None:
        if start == len(s):
            result.append(path)
        
        for word in wordDict:
            n = len(word)
            if start + n > len(s):
                continue

            if s[start: start+n] == word:
                self.wordBreakRec(s, path + (word if path == "" else " " + word), result, wordDict, start+n) 

    def wordBreakTree(self, s: str, wordDict: List[str]) -> Dict:
        n = len(s)
        parents = defaultdict(list)
        leaves = set()
        
        queue = deque()

        # (word, start, depth)
        queue.append(("", 0, 0))
        while queue:
            word, word_start, depth = queue.popleft()
            word_end = word_start + len(word)
            if word_end == n:
                leaves.add(word)
                continue
            
            for child in wordDict:
                m = len(child)
                child_start = word_end
                child_end = child_start + m
                if child_end > n:
                    continue
                if s[child_start: child_end] == child:
                    if parents[child_start] == []:
                        parents[child_start] = set()
                    parents[child_start].add(word)
                    queue.append((child, child_start, depth+1))
        
        #print("leaves:\n", leaves)
        #print("parents:\n", parents)

        result = []
        def dfs(node, idx, path):
            if node == "":
                result.append(path)
                return
            for parent in parents[idx]:
                dfs(parent, idx - len(parent), (node + " " + path) if path != "" else node)

        for leaf in leaves:
            dfs(leaf, n - len(leaf), "")

        return result

    def wordBreak(self, s: str, wordDict: List[str]) -> List[str]:
        result = self.wordBreakTree(s, wordDict)
        return result

        # result = []
        # self.wordBreakRec(s, "", result, wordDict, 0)
        # result = list(set(result))
        # return result
```

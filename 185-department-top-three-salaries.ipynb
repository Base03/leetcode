```python
import pandas as pd

def top_three_salaries(employee: pd.DataFrame, department: pd.DataFrame) -> pd.DataFrame:
    # Rename columns
    department=department.rename(columns={'id':'dept_id','name':'Department'})
    employee=employee.rename(columns={'name':'Employee','salary':'Salary'})

    # merge dataframes and calculate ranks
    merged_df=employee.merge(department, left_on='departmentId', right_on='dept_id',how='left')
    merged_df['rnk']= merged_df.groupby('Department')['Salary'].rank(method='dense',ascending=False)
    cond_1=merged_df['rnk']<=3

    soln_df = merged_df.loc[cond_1, ['Department', 'Employee', 'Salary']]
    
    return soln_df
```

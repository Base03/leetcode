```python
from typing import List
from typing import Deque
class Solution:
    def simplifyPath(self, path: str) -> str:
        # split on /
        parts = path.split('/')
        stack: Deque[str] = Deque()
        for part in parts:
            if part == '' or part == '.':
                continue
            elif part == '..':
                if stack:
                    stack.pop()
            else:
                stack.append(part)
                
        return '/' + '/'.join(stack)
        
```

```python
soln = Solution()

path = "/home/"
expected = "/home"
print(f"simplified path: {soln.simplifyPath(path)}, expected: {expected}, equality: {soln.simplifyPath(path) == expected}")

path = "/home//foo/"
expected = "/home/foo"
print(f"simplified path: {soln.simplifyPath(path)}, expected: {expected}, equality: {soln.simplifyPath(path) == expected}")

path = "/home/user/Documents/../Pictures"
expected = "/home/user/Pictures"
print(f"simplified path: {soln.simplifyPath(path)}, expected: {expected}, equality: {soln.simplifyPath(path) == expected}")

path = "/../"
expected = "/"
print(f"simplified path: {soln.simplifyPath(path)}, expected: {expected}, equality: {soln.simplifyPath(path) == expected}")

path = "/.../a/../b/c/../d/./"
expected = "/.../b/d"
print(f"simplified path: {soln.simplifyPath(path)}, expected: {expected}, equality: {soln.simplifyPath(path) == expected}")
```

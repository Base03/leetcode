```python
class Solution:
    # returns true if s is a valid number
    # basically, just try to parse s as a float and see if it throws an exception
    # if it doesn't, check the edge cases where the float is infinity or NaN and return
    # False since this method isn't supposed to handle those cases
    def isNumber(self, s: str) -> bool:
        try:
            float(s)
            if "inf" in s.lower() or "nan" in s.lower() or "infinity" in s.lower():
                return False
            return True
        except:
            return False
        

    dig = "0123456789"
    sgn = "+-"
        
    # returns true if s is a valid sequence of digits
    def isDigits(self, s: str) -> bool:
        if not s:
            return False
        for char in s:
            if char not in self.dig:
                return False
        return True
            
    # returns true if s is a valid integer
    def isInteger(self, s: str) -> bool:
        if not s:
            return False
        
        if s[0] in self.sgn:
            s = s[1:]
        return self.isDigits(s)
    
    # returns true if s is a valid decimal number or a valid integer
    def isDecimal(self, s: str) -> bool:
        if not s or s.count(".") != 1:
            return self.isInteger(s)
        
        if s[0] in self.sgn:
            s = s[1:]

        s1, s2 = s.split(".")
        v1, v2 = self.isDigits(s1), self.isDigits(s2)
        return ((v1 and v2) or (not s1 and v2) or (v1 and not s2))

    # returns true if s is a valid number
    # first checks if s is empty and returns False if it is
    # then converts s to lowercase so that "E" -> "e", Infinity -> infinity, etc.
    # then checks if s is in exponential notation by checking if there is a single "e"
    # in s and if so, splits on "e" and checks if the first part is a valid decimal
    # and the second part is a valid integer
    # otherwise checks if s is a valid decimal
    def isNumber(self, s: str) -> bool:
        if not s:
            return False
        s = s.lower()

        if "e" in s and s.count("e") == 1:
            s1, s2 = s.split("e")
            return self.isDecimal(s1) and self.isInteger(s2)
        else:
            return self.isDecimal(s)
```

```python
# test 1
s = ""
print(Solution().isNumber(s), "expected:", False)
s = " "
print(Solution().isNumber(s), "expected:", False)
s = "+"
print(Solution().isNumber(s), "expected:", False)
s = "-"
print(Solution().isNumber(s), "expected:", False)
s = "."
print(Solution().isNumber(s), "expected:", False)
s = "e"
print(Solution().isNumber(s), "expected:", False)
s = "E"
print(Solution().isNumber(s), "expected:", False)
s = "1"
print(Solution().isNumber(s), "expected:", True)
s = "1e"
print(Solution().isNumber(s), "expected:", False)
s = "1e1"
print(Solution().isNumber(s), "expected:", True)
s = "1e+1"
print(Solution().isNumber(s), "expected:", True)
s = "1e-1"
print(Solution().isNumber(s), "expected:", True)
s = "1.1"   
print(Solution().isNumber(s), "expected:", True)
```

```python
class Solution:
    # O(S*J)
    def numJewelsInStones(self, jewels: str, stones: str) -> int:
        num_jewels: int = 0
        for stone in stones:
            if stone in jewels:
                num_jewels += 1
        return num_jewels
```

```python
from typing import Set

class Solution:
    # O(J + S + s) = O(J + S) just probably slower than the previous loop version
    def numJewelsInStones(self, jewels: str, stones: str) -> int:
        num_jewels: int = 0
        # O(J)
        jewels: Set = set(jewels)
        print(f"jewels: {jewels}")

        # O(S)
        stones_dict: dict[str, int] = {}
        for stone in stones:
            stones_dict[stone] = stones_dict.get(stone, 0) + 1
        
        # redundant relative to previous version because the lookup is still O(1)
        # but better if we kept collecting stones and had persistent data storage
        # O(s) where s is unique keys in the dict
        for stone, stone_count in stones_dict.items():
            if stone in jewels:
                num_jewels += stone_count
        
        return num_jewels
```

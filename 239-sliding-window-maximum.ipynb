```python
from typing import List
from functools import cache

class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        n: int = len(nums)
        max_window: List[int] = [0] * (n - k + 1)

        """
        i and j are inclusive
        0 <= i <= j < n
        add memoization flag
        """
        @cache
        def max_in_range(i: int, j: int) -> int:
            if i == j:
                return nums[i]
            else:
                return max(nums[j], max_in_range(i, j-1))
            
        for i in range(n - k +1):
            max_window[i] = max_in_range(i, i+k-1)

        return max_window
```

```python
from typing import Tuple, List

class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        n: int = len(nums)
        max_window: List[int] = [0] * (n - k + 1)

        def max_index(i: int, j: int) -> Tuple[int, int]:
            max_val = nums[i]
            max_idx = i

            for idx in range(i+1, j):
                if nums[idx] >= max_val:
                    max_val = nums[idx]
                    max_idx = idx
            
            return max_idx, max_val
        
        max_idx, max_val = max_index(0, k)
        max_window[0] = max_val
        for idx in range(k, n):
            #print(f"idx: {idx}, max_window: {max_window}")
            if nums[idx] >= max_val:
                max_val = nums[idx]
                max_idx = idx
            elif max_idx <= idx - k:
                max_idx, max_val = max_index(max_idx+1, idx+1)

            max_window[idx-k+1] = max_val

        return max_window
```

```python
from collections import deque
from typing import List

class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        dq = deque()
        n = len(nums)
        result = [0]* (n - k + 1)
        idx = 0
        

        for i in range(n):
            if dq and dq[0] == i - k:
                dq.popleft()
            
            while dq and nums[dq[-1]] < nums[i]:
                dq.pop()

            dq.append(i)
            if i >= k - 1:
                result[idx] = nums[dq[0]]
                idx += 1

        return result
```

```python
soln = Solution()

input = [1,3,-1,-3,5,3,6,7]
k = 3
expected = [3,3,5,5,6,7]
output = soln.maxSlidingWindow(input, k)
print(f"output: {output}, expected: {expected}, pass: {output == expected}")

input = [1, -1]
k = 1
expected = [1, -1]
output = soln.maxSlidingWindow(input, k)
print(f"output: {output}, expected: {expected}, pass: {output == expected}")

input = [7, 2, 4]
k = 2
expected = [7, 4]
output = soln.maxSlidingWindow(input, k)
print(f"output: {output}, expected: {expected}, pass: {output == expected}")
```

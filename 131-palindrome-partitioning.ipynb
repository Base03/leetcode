```python
from typing import List

class Solution:
    def partition(self, s: str) -> List[List[str]]:
        result = []
        n = len(s)

        def dfs(i: int, partition: List[str]):
            if i == n:
                result.append(partition[:])
            
            for j in range(i,n):
                piece = s[i:j+1]
                if piece == piece[::-1]:
                    partition.append(piece)
                    dfs(j+1, partition)
                    partition.pop()
        
        dfs(0,[])
        return result
```

```python
soln = Solution()

s = "abba"
result = soln.partition(s)
print(f"result: {result}")
```

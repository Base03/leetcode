```python
from typing import List

class Solution:
    def countTrapezoids(self, points: List[List[int]]) -> int:
        slope_dict: dict[tuple[int, int], List[tuple[int, int]]] = {}
        # hash all pairs of points into lists with the same slope. need to normalize dy/dx via gcd
        for i, a_point in enumerate(points):
            for j, b_point in enumerate(points[:i]):
                normalized_slope = self.gcd((a_point[0] - b_point[0], a_point[1] - b_point[1]))
                if normalized_slope not in slope_dict:
                    slope_dict[normalized_slope] = []
                slope_dict[normalized_slope].append((i, j))

        print(slope_dict)
        return 0

    def gcd(self, dxdy: tuple[int, int]) -> tuple[int, int]:
        # input is (dx, dy), return normalized (dx, dy) such that the greatest common divisor of dx and dy is 1
        
        dx, dy = dxdy
        if dy == 0:
            return (1, 0)
        if dx == 0:
            return (0, 1)
        sign = 1
        if dx * dy < 0:
            sign = -1
        dx = abs(dx)
        dy = abs(dy)
        while dy:
            dx, dy = dy, dx % dy
        gcd = dx
        return (sign * dxdy[0] // gcd, abs(dxdy[1]) // gcd)
        
```

```python
soln = Solution()

points: List[List[int]] = [[0,0], [1,1], [0, -1], [1,0]]
soln.countTrapezoids(points)
```

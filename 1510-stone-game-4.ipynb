```python
import sys
sys.setrecursionlimit(10**8)

class Solution:
    def winnerSquareGame(self, n: int) -> bool:
        dp = {}

        def helper(n: int) -> int:
            # if n is already at 0, then the current player has lost
            if n == 0:
                return -1
            
            # if n is already in the dp, then return the value
            if n in dp:
                return dp[n]
            
            base = n**0.5 // 1
            current = -1
            # check all possible moves
            while base * base >= 1:
                # score is 1 if the next player loses
                current = max(current, -helper(n-base*base))
                if current == 1:
                    break
                base -= 1
            dp[n] = current
            return current
        
        return helper(n) == 1

    
```

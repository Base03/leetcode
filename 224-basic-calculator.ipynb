```python
class Solution:
    def calculate(self, s: str) -> int:
        # calculator consists only of digits, '+', '-' and ' '
        # strip all spaces
        s = s.replace(' ', '')
        # replace all '-' with '+-' to split by '+'
        s = s.replace('-', '+-')
        splits= s.split('+')

        accumulator = 0
        for num in splits:
            accumulator += int(num)
        return accumulator
```

```python
soln = Solution()

input = "1 + 1"
expected = 2
output = soln.calculate(input)
print(f"expected: {expected}, output: {output}, equal: {expected == output}")

input = " 2-1 + 2 "
expected = 3
output = soln.calculate(input)
print(f"expected: {expected}, output: {output}, equal: {expected == output}")

input = "1+4+5+2-3+6+8"
expected = 23
output = soln.calculate(input)
print(f"expected: {expected}, output: {output}, equal: {expected == output}")
```

```python
# improved calculator where we can have parentheses. If it were only addition, the parentheses wouldn't 
# matter, but subtraction permits negation of entire sub-expressions

class Solution:
    def calculate(self, s: str) -> int:
        num = 0
        stack = []
        result = 0
        sign = 1  # 1 means positive, -1 means negative
        for char in s:
            if char.isdigit():
                num = num * 10 + int(char)
            elif char in "+-":
                result += sign * num
                num = 0
                sign = 1 if char == "+" else -1
            elif char == "(":
                stack.append(result)
                stack.append(sign)
                result = 0
                sign = 1
                num = 0
            elif char == ")":
                result += sign * num
                num = 0
                result *= stack.pop() 
                result += stack.pop()  

        result += sign * num
        return result
```

```python
sol = Solution()
s = "1 + 1"
expected = 2
output = sol.calculate(s)
print(f"expected: {expected}, output: {output}, equal: {expected == output}")
```

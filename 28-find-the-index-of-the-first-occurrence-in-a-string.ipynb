```python
# easy brute force

class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        nN: int = len(needle)
        nH: int = len(haystack)

        for i in range(nH - nN + 1):
            for j in range(nN):
                if haystack[i + j] != needle[j]:
                    break
            else:
                return i

        return -1
```

```python
# KMP (Kunth-Morris-Pratt)

class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        nN: int = len(needle)
        nH: int = len(haystack)

        if needle == "":
            return 0
        
        lps = [0] * nN
        i = 1
        prevLPS = 0
        while i < nN:
            if needle[i] == needle[prevLPS]:
                lps[i] = prevLPS+1
                prevLPS += 1
                i += 1
            elif prevLPS == 0:
                lps[i] = 0
                i += 1
            else:
                prevLPS = lps[prevLPS - 1]

        i = 0 # pointer for haystack
        j = 0 # pointer for needle
        while i < nH:
            if haystack[i] == needle[j]:
                i += 1
                j += 1
            else:
                if j == 0:
                    i += 1
                else:
                    j = lps[j-1]

            if j == nN:
                return i - nN
        return -1
```

```python
# KMP alternate

class Solution:
    def  strStr(self, haystack: str, needle: str) -> int:
        nN, nH = len(needle), len(haystack)
        lps = [0]* nN

        # build lps table:
        prefix = 0
        for i in range(1, nN):
            while prefix > 0 and needle[i] != needle[prefix]:
                prefix = lps[prefix-1]
```

```python
soln = Solution()

haystack = "sadbutsad"
needle = "sad"
expected = 0
result = soln.strStr(haystack, needle)
print(f"expected: {expected}, result: {result}")

haystack = "kittysadbutsad"
needle = "sad"
expected = 5
result = soln.strStr(haystack, needle)
print(f"expected: {expected}, result: {result}")

haystack = "leetcode"
needle = "leeto"
expected = -1
result = soln.strStr(haystack, needle)
print(f"expected: {expected}, result: {result}")
```

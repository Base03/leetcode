```python
from typing import List

class Solution:
    def isScrambleMemo(self, s1: str, s2: str) -> bool:
        memo = {}
        result = False

        def helper(strings: List[str], depth) -> bool:
            
            string_key = ".".join(strings)
            # print(" "*depth, string_key)

            if string_key in memo:
                return memo[string_key]
            if "".join(strings) == s2:
                return True
            else:
                result = False
                # filter out strings which have characters irrevocably different from s2
                # once a substring has length 1, it must be equal to the corresponding location
                # in s2. If not, we can discard this branch
                count = 0
                for i in range(len(strings)):
                    if len(strings[i]) == 1:
                        if strings[i][0] != s2[count]:
                            memo[string_key] = False
                            return memo[string_key]
                    count += len(strings[i])

                for i in range(len(strings)):
                    for j in range(1, len(strings[i])):
                        sub_str1 = strings[i][:j]
                        sub_str2 = strings[i][j:]

                        new_strings = strings[:i] + [sub_str1] + [sub_str2] + strings[i+1:]
                        # print(" "*depth, new_strings)
                        result = result or helper(new_strings, depth+1)
                        
                        new_strings = strings[:i] + [sub_str2] + [sub_str1] + strings[i+1:]
                        # print(" "*depth, new_strings)
                        result = result or helper(new_strings, depth+1)
            
                memo[string_key] = result
                return memo[string_key]
        
        return helper([s1], 0)
    
    def isScrambleEfficientMemo(self, s1: str, s2: str) -> bool:
        memo = {}
        # check to see if s1 and s2 can be formed from the
        # substring + reverse operation
        def helper(s1: str, s2: str):
            if s1 == s2:
                return True
            
            if (s1, s2) in memo:
                return memo[(s1, s2)]
            
            if len(s1) != len(s2) or sorted(s1) != sorted(s2):
                return False
            
            n = len(s1)
            for k in range(1, n):
                if helper(s1[:k], s2[:k]) and helper(s1[k:], s2[k:]):
                    memo[(s1, s2)] = True
                    return True
                if helper(s1[:k], s2[-k:]) and helper(s1[k:], s2[:-k]):
                    memo[(s1, s2)] = True
                    return True
            memo[(s1, s2)] = False
            return False
            
        return helper(s1, s2)
    
    def isScramble(self, s1: str, s2: str) -> bool:
        return self.isScrambleEfficientMemo(s1, s2)
        #return self.isScrambleMemo(s1, s2)
```

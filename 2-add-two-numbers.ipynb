```python
#Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

    def __repr__(self):
        return f"ListNode(val={self.val}, next={self.next})"
    
    # string representation for easier visualization
    def __str__(self):
        result = []
        current = self
        while current:
            result.append(str(current.val))
            current = current.next
        return " -> ".join(result)
    
```

```python
from typing import Optional, List
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        
        def recurse(node1, node2, parent, carry_in) -> int:
            if node1 is not None or node2 is not None or carry_in > 0:
                node1_val = node1.val if node1 is not None else 0
                node2_val = node2.val if node2 is not None else 0

                remainder = (node1_val + node2_val + carry_in) % 10
                carry = (node1_val + node2_val + carry_in) // 10
                child = ListNode(remainder)
                parent.next = child
                
                recurse(node1.next if node1 is not None else None, 
                        node2.next if node2 is not None else None, 
                        child, carry)
                

        dummy = ListNode(0)
        recurse(l1,l2,dummy,0)
        return dummy.next
```

```python
from typing import Optional, List
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:

        dummy = ListNode()
        temp = dummy
        carry = 0

        while l1 or l2 or carry:
            val1 = l1.val if l1 else 0
            val2 = l2.val if l2 else 0

            net = val1 + val2 + carry
            carry = net // 10
            val = net % 10

            temp.next = ListNode(val)
            temp = temp.next

            if l1:
                l1 = l1.next
            if l2:
                l2 = l2.next

        return dummy.next
```

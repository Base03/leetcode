```python
from typing import List
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        if not nums or len(nums) == 0:
            return 0
        
        ptr: int = 1
        count: int = 1

        for i in range(1, len(nums)):
            if nums[i] == nums[i-1]:
                count += 1
            else:
                count = 1
            
            if count <= 2:
                nums[ptr] = nums[i]
                ptr += 1
        return ptr
```

```python
soln = Solution()
nums = [1,1,1,2,2,3]
count = soln.removeDuplicates(nums)
print(f"nums: {nums}, len(nums): {len(nums)}, count: {count}")
```

```python
from collections import defaultdict


class Solution:
    def minWindow(self, s: str, t: str) -> str:
        count_dict = {}
        for char in t:
            if char in count_dict:
                count_dict[char] += 1
            else:
                count_dict[char] = 1
        
        #print(count_dict)

        idx_dict = {}
        for i in range(len(s)):
            char = s[i]
            if char in count_dict:
                if char not in idx_dict:
                    idx_dict[char] = [i]
                else:
                    idx_dict[char].append(i)

        print(idx_dict)
        return ''
    
    def minWindow(self, s: str, t: str) -> str:
        t_counts = defaultdict(int)
        for char in t:
            t_counts[char] += 1
        
        matched = 0
        total = len(t_counts.keys())
        l_ptr = 0
        r_ptr = 0
        window_dict = defaultdict(int)
        match = [-1, -1]
        
        # march right with the right most pointer
        while r_ptr < len(s):
            if s[r_ptr] in t_counts:
                # increment counts for character in the window
                window_dict[s[r_ptr]] += 1
                # check if we have matched sufficient counts for the character
                if window_dict[s[r_ptr]] == t_counts[s[r_ptr]]:
                    matched += 1

            # if we have matched all characters, march left pointer to minimize window size
            while l_ptr <= r_ptr and matched == total:
                # if we have a smaller window, update the match
                if (match[0] == -1) or (r_ptr - l_ptr < match[1] - match[0]):
                    match = [l_ptr, r_ptr]
                    
                if s[l_ptr] in t_counts:
                    # if we have decremented the count to less than the required count
                    # decrement the matched count and we will have to march with the right pointer
                    if window_dict[s[l_ptr]] == t_counts[s[l_ptr]]:
                        matched -= 1
                    # decrement counts for character in the window
                    window_dict[s[l_ptr]] -= 1
                
                # increment the left pointer
                l_ptr += 1

            # increment the right pointer
            r_ptr += 1

        if match[0] != -1:
            return s[match[0]:match[1]+1]
        return ""
```
